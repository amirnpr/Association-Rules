```python
import pandas as pd
pd.set_option('display.width', 100)
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', 12)
from mlxtend.preprocessing import TransactionEncoder 
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import fpgrowth
```

# Read Data - Transaction Table

```python
file = pd.read_excel(r"C:\Users\Lenovo\Dropbox\My PC (DESKTOP-BHVANOA)\Desktop\Market.xlsx")

lista = file.to_numpy().tolist()

listb = [[] * 0 for i in range(len(lista))]

for i in range(len(lista)):
    listb[i] = [item.split(',') for item in lista[i]]

dataset = []
for i in range(len(lista)):
    dataset.append(listb[i][0])


number_of_transactions = len(dataset)

pd.DataFrame(dataset)
```

```python
te = TransactionEncoder()
te_array = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_array, columns=te.columns_)
df
```

# Find Frequent Itemsets

```python
frequent_itemsets_fp = fpgrowth(df, min_support=0.03, use_colnames=True)	
frequent_itemsets_fp
```

# Association Rules

```python
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.3)

del rules_fp["leverage"]
del rules_fp["conviction"]
rules_fp
```

# Convert DataFrame to List

```python
list1 = rules_fp.values.tolist()
# print(list1)
number_of_rules = len(list1)
```

# Calculate P(A|B) & P(B|A)

P(B|A) = Support (AUB) / Support (A) = Confidence (A=>B) 

P(A|B) = Support (AUB) / Support (B)

```python
table_2 = [[0] * (2) for i in range(number_of_rules)]

for i in range(number_of_rules):
    table_2[i][0] = round(list1[i][5], 3)                        #P(B|A)
    table_2[i][1] = round(list1[i][4] / list1[i][3], 3)        #P(A|B)

pd.DataFrame(table_2, columns=['P(B|A)', 'P(A|B)'])
```

# Evaluation Measures

All Confidence(A,B) = Min {P(A|B),P(B|A)}

MAX Confidence(A,B) = Max {P(A|B),P(B|A)}

Kulczynski(A,B) =  P(A|B)+P(B|A) / 2

Cosine(A,B) = sqrt(P(A|B)*P(B|A))

Imbalance Ratio(A,B) = |sup(A) - sup(B)| / sup(A) + sup(B) - sup(AUB)

```python
table_4 = [[0] * (5) for i in range(number_of_rules)]

from math import sqrt

for i in range(number_of_rules):
    table_4[i][0] = round(min(table_2[i]), 3)                             #All Confidence
    table_4[i][1] = round(max(table_2[i]), 3)                             #MAX Confidence
    table_4[i][2] = round((table_2[i][0] + table_2[i][1]) / 2, 3)         #Kulczynski     
    table_4[i][3] = round(sqrt(table_2[i][0] * table_2[i][1]), 3)         #Cosine
    table_4[i][4] = round(abs(list1[i][2] - list1[i][3]) / (list1[i][2] + list1[i][3] - list1[i][4]), 3)   #IR

pd.DataFrame(table_4, columns=['All Confidence', 'MAX Confidence', 'Kulczynski', 'Cosine', 'Imbalance Ratio (IR)'])
```

# Chi-Square 

```python
def chi_square(supAUB,supA,supB,n):

    contingency_table = [[0] * (3) for i in range(3)]

    contingency_table[0][0] = round(supAUB * n)
    contingency_table[1][0] = round((supA - supAUB) * n)
    contingency_table[2][0] = round((supA * n))
    contingency_table[0][1] = round((supB - supAUB) * n)
    contingency_table[0][2] = round((supB * n))
    contingency_table[2][1] = round((1-supA) * n)
    contingency_table[1][2] = round((1-supB) * n)
    contingency_table[1][1] = round((1-supA-supB+supAUB) * n)
    contingency_table[2][2] = n

    X2 = 0

    expected_table = [[0] * (2) for i in range(2)]

    for i in range(2):
        for j in range(2):
            expected_table[i][j] = (contingency_table[-1][j] * contingency_table[i][-1]) / n
            X2 += (pow((expected_table[i][j] - contingency_table[i][j]), 2)) / expected_table[i][j]
    
    return round(X2)

table_3 = [[0] * (1) for i in range(number_of_rules)]

for i in range(number_of_rules):
    table_3[i][0] = chi_square(list1[i][4], list1[i][2], list1[i][3], number_of_transactions)   #Calculate Chi-Square
    list1[i].append(table_3[i][0])                                                      #Add Chi-Square Values

pd.DataFrame(list1, columns =['LHS', 'RHS','LHS support','RHS support','Support','Confidence','Lift','Chi-Square'])
```

# Final Table

```python
for i in range(number_of_rules):
    list1[i].append(table_4[i][0])
    list1[i].append(table_4[i][1])
    list1[i].append(table_4[i][2])
    list1[i].append(table_4[i][3])
    list1[i].append(table_4[i][4])

for i in range(number_of_rules):
    list1[i][2] = round(list1[i][2], 3)
    list1[i][3] = round(list1[i][3], 3)
    list1[i][4] = round(list1[i][4], 3)
    list1[i][5] = round(list1[i][5], 3)
    list1[i][6] = round(list1[i][6], 3)

final_table = pd.DataFrame(list1 , columns =['LHS', 'RHS','LHS support','RHS support',
'Support','Confidence','Lift','Chi-Square','All_Confidence',
'Max_Confidence','Kulczynski','Cosine','Imbalance Ratio (IR)']) 

final_table
```
